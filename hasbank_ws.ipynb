```
from bs4 import BeautifulSoup
import requests
import pandas as pd
from requests_html import HTMLSession
from datetime import datetime
```

```
#has bank news

try:
    url = requests.get('https://www.xacbank.mn/news?lang=en')
    soup = BeautifulSoup(url.text, 'html.parser')
    articles = soup.find('div', class_ = 'maintext')
    
    rows = []
    for article in articles.find_all('div', class_='uk-width-3-5'):
        title = article.find('h4' , class_ ="title").text
        date = article.find('span').text.lstrip()
        link = article.a['href']
        description = article.find('p').text
        category = 'Finance'
       
        row_dict = {'Title':title, 'Date':date, 'Link':link, 'Category':category, 'Description':description} 
        rows.append(row_dict)
        
        has_bank = pd.DataFrame(rows)
        
except Exception as e:
    print(e)
```

```
has_bank['Date'] = has_bank['Date'].apply(lambda x: datetime.strptime(x,'%d %b, %Y'))
```

```
has_bank.to_csv('/datasets/webscrape/has_bank_'+ datetime.now().strftime('%m_%d_%y.csv'), index_label = 'id')
```
