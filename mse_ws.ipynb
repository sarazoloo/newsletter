```
import feedparser 
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
from requests_html import HTMLSession
from datetime import datetime
```

```
#mse rss feed
m = feedparser.parse('https://www.mse.mn/en/feed')

try:
    rows = []
    for j in range(1, len(m.entries)):
        title = m.entries[j].title
        link = m.entries[j].link
        date = m.entries[j].published
        description = m.entries[j].description
        category = 'Finance'
        
        row_dict = {'Title':title, 'Date':date, 'Link':link, 'Category':category, 'Description':description}
        rows.append(row_dict)
        
        mse = pd.DataFrame(rows)
        
except Exception as e:
    print(e)
```

```
mse['Date'] = mse['Date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S'))

```

```
mse['Date'] = mse['Date'].apply(lambda x: datetime.strftime(x, '%Y-%m-%d'))
```

```
mse.to_csv('/datasets/webscrape/mse_'+ datetime.now().strftime('%m_%d_%y.csv'), index_label = 'id')
```
