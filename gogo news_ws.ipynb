```
from bs4 import BeautifulSoup
import requests
import pandas as pd
from requests_html import HTMLSession
from datetime import datetime
```

```
try:
    url = "https://mongolia.gogo.mn/"
    s = HTMLSession()
    r = s.get(url)

    categories = r.html.xpath('//*[@id="gogoapp"]/div[1]/section/div[2]/div/nav/div[1]/ul', first = True)

    rows = []
    for category in categories.absolute_links:
        r = s.get(category)
        pages = category+'/fetch'
    
        if pages != 'https://mongolia.gogo.mn//fetch':
            url = requests.get(pages)
            content = url.json()
        
        
            for item in content['nav_news_list']:
                title = item['title']
                description = item['head_line'].strip()
                date = item['publish_date'].split('T', 1)[0]
                category = item['newstyppe']['news_type_name']
                lid = item['lid']
    
                link = 'https://mongolia.gogo.mn/r/'+lid
        

                row_dict = {'Title':title, 'Date':date, 'Link':link, 'Category':category, 'Description':description}
                rows.append(row_dict)
        
                gogo = pd.DataFrame(rows)

except Exception as e:
    print(e)                
```

```
gogo['Date'] = gogo['Date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
```

```
gogo.to_csv('/datasets/webscrape/gogo_'+ datetime.now().strftime('%m_%d_%y.csv'), index_label = 'id')
```
