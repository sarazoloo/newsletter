```
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
from datetime import datetime, timedelta
```

```
#montsame webscrape
try:
    html = requests.get('https://montsame.mn/en/')
    soup = BeautifulSoup(html.text, 'html.parser')
    pages = soup.find('ul', class_ = 'navbar-nav mr-auto montsame-nav')

    rows = []
    for page in pages.find_all('li', class_ ='nav-item' ):
        links = 'https://montsame.mn' + page.a['href']
        topic = page.a.text
    
        category = "".join([s for s in topic.strip().splitlines(True) if s.strip()])
    
        if links != 'https://montsame.mn#' and 'https://montsame.mn/en/':
            url = requests.get(links)
            soup1 = BeautifulSoup(url.text, 'html.parser')
            articles = soup1.find('div', class_ = 'row no-gutters')
        
        
            for article in articles.find_all('div', class_ = 'news-box grid col-sm-6 p-2'):
                link = 'https://montsame.mn' + article.a['href']
                title = article.find('div', class_ = 'title content-en').text
                description = article.find('div', class_ = 'body content-en').text.replace(u'\xa0', u' ')
                date_published = article.find('div', class_ ="stat d-block").text.strip().split()[0]
    
    
                row_dict = {'Title':title, 'Date':date_published, 'Link':link, 'Category':category, 'Description':description}
                rows.append(row_dict)
        
                montsame = pd.DataFrame(rows)
    
    

except Exception as e:
    print(e)
    
```

```
montsame['Category'] = montsame['Category'].replace('Economy', 'Economics')
```

```
montsame['Date'] = montsame['Date'].astype(int)
```

```
for num in range(0, len(montsame['Date'])):
    u = datetime.now() - timedelta(days = int(montsame['Date'][num]))
    montsame['Date'][num] = u.strftime('%Y-%m-%d')
```
