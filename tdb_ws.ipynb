```
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
from datetime import datetime
```

```
try: 
    url = requests.get('https://www.tdbsecurities.mn/category/other-news?lang=en')
    soup = BeautifulSoup(url.text, 'html.parser')
    articles = soup.find('div', class_ = 'uk-grid uk-grid-medium') 

    rows = []
    for article in articles.find_all('div', class_ = 'uk-width-1-1 uk-width-small-1-2 uk-width-large-1-4'):
        link = article.a['href']
        title = article.find('h3').text
        date = article.find_all('p')[0].text.replace('\xa0','').lstrip()
        description = article.find_all('p')[1].text
        category = 'Finance'

        a , b = date.split('• ',1)
        Date = date[:4] + '/' + date[4:]
    
        row_dict = {'Title':title, 'Date':a, 'Link':link, 'Category':category, 'Description':description}
        rows.append(row_dict)
        
        tdb = pd.DataFrame(rows)
    
        
except Exception as e:
    print(e)
```

```
tdb['Date'] = tdb['Date'].apply(lambda x: datetime.strptime(x,'%Y.%m.%d'))
```

```
tdb.to_csv('/datasets/webscrape/tdb_' + datetime.now().strftime('%m_%d_%y.csv'), index_label='id')
```
