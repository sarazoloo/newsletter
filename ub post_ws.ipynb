```
import feedparser 
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
from datetime import datetime
from requests_html import HTMLSession
```

```
#ubpost rss
try:
    html = requests.get('https://theubposts.com/')
    soup = BeautifulSoup(html.text, 'html.parser')
    pages = soup.find('ul', class_ = 'hide-on-med-and-down')

    rows = []
    for page in pages.find_all('li'):
        links = 'https://theubposts.com' + page.a['href']
        topic = page.a.text
        category = "".join([s for s in topic.strip().splitlines(True) if s.strip()])
    
        if links != 'https://theubposts.com/' and 'https://theubposts.com#!':
            url = requests.get(links)
            soup1 = BeautifulSoup(url.text, 'html.parser')
            articles = soup1.find('div', class_ = 'category_wrap')
        
            if articles is not None:
                for article in articles.find_all('div', class_ = 'col s12 m6 l6'):
                    link = 'https://theubposts.com' + article.a['href']
                    title = article.find('h3', class_ = 'module_title').text
                    date = article.find('li', class_ ="date").text

                    row_dict = {'Title':title, 'Date':date, 'Link':link, 'Category':category}
                    rows.append(row_dict)
                
                    ub_post = pd.DataFrame(rows)
    
except Exception as e:
    print(e)
```

```
ub_post['Date'] = ub_post['Date'].apply(lambda x: datetime.strptime(x,'%b %d,%Y'))
        
```

```
 ub_post.to_csv('/datasets/webscrape/ub_post_' + datetime.now().strftime('%m_%d_%y.csv'), index_label='id')
```
