```
import feedparser 
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
from csv import writer
from datetime import datetime
from requests_html import HTMLSession
```

```
#google alerts rss
try:
    
    feed = feedparser.parse('https://www.google.com/alerts/feeds/12655880066555305842/4740245944062060987')
    
    rows = []
    for f in range(0,len(feed.entries)):
        title = BeautifulSoup(feed.entries[f].title, "html.parser").text
        date = feed.entries[f].published.split('T')[0]
        link = feed.entries[f].link
        description = BeautifulSoup(feed.entries[f].summary, "html.parser").text
        category = 'International'
   
        row_dict = {'Title':title, 'Date':date, 'Link':link, 'Category':category, 'Description':description }
        rows.append(row_dict)

        google_alerts = pd.DataFrame(rows)

        
except Exception as e:
    print(e)
```

```
google_alerts['Date'] = google_alerts['Date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
```

```
google_alerts.to_csv('/datasets/webscrape/google_alert_'+ datetime.now().strftime('%m_%d_%y.csv'), index_label = 'id')
```
