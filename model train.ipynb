```
import pandas as pd
import numpy as np
from sklearn import preprocessing
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn import metrics
import xgboost as xgb
from sklearn import *
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer
import pickle
import joblib
import os
from sklearn.linear_model import LogisticRegression
```

```
df = pd.read_csv('merged.csv')
```

```
df.head()
```

```
df = df.drop(columns=['Date', 'Link', 'Category'])
```

```
df["Text"] = df['Title'] + ': ' + df["Description"]
```

```
df = df.drop(columns=['Title', 'Description'])
```

```
df = df.rename(columns={'Send':'y', 'Text':'X'})
```

```
df
```

```
X = df['X']
```

```
X
```

```
y = df['y']
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0, stratify=y)
```

```
vectorizer = feature_extraction.text.CountVectorizer()
```

```
count_vect = vectorizer.fit_transform(X_train)
```

```
test_count_vect = vectorizer.transform(X_test)
```

```
lr = LogisticRegression()
lr.fit(count_vect, y_train)
lr_pred = lr.predict(test_count_vect)

#lr_pred_proba = lr.predict_proba(test_count_vect)
#print(lr_pred_proba)
```

```
metrics.accuracy_score(y_test, lr_pred)
```
