```
import feedparser 
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
from datetime import datetime
```

```
#news.mn economic web scrape
try:
    g = feedparser.parse('https://news.mn/en/feed/')
    

    rows = []
    for i in range(0,len(g.entries)):
        title = g.entries[i].title
        link = g.entries[i].link
        date = g.entries[i].published
        category = g.entries[i].category
        description = BeautifulSoup(g.entries[i].summary, "html.parser").text

        row_dict = {'Title':title, 'Date':date, 'Link':link, 'Description':description, 'Category':category}
        rows.append(row_dict)

        news_mn = pd.DataFrame(rows)
        
        
except Exception as e:
    print(e)
```

```
news_mn['Category'] = news_mn['Category'].replace('Economic', 'Economics')
```

```
news_mn['Date'] = news_mn['Date'].apply(lambda x: datetime.strptime(x,'%a, %d %b %Y %H:%M:%S %z'))
```

```
news_mn['Date'] = news_mn['Date'].apply(lambda x: datetime.strftime(x, '%Y-%m-%d'))
```

```
news_mn.to_csv('/datasets/webscrape/news_mn_' + datetime.now().strftime('%m_%d_%y.csv'), index_label='id')
```
