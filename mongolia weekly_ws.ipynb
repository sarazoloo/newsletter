```
import feedparser 
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
from datetime import datetime
import re
```

```
#mongolia_weekly rss
try:
    d = feedparser.parse('https://www.mongoliaweekly.org/blog-feed.xml')
    
    rows = []
    for n in range(0,len(d.entries)):
        if "tags" in d.entries[n].keys():
            tags = [x.term for x in d.entries[n].tags][0]
        else:
            tags = 'Economics'
        title = d.entries[n].title
        date = d.entries[n].published.split(',')[1].lstrip()
        link = d.entries[n].link
        description = d.entries[n].summary
   
        row_dict = {'Title':title, 'Date':date, 'Link':link, 'Category':tags, 'Description':description}
        rows.append(row_dict)

        mgl_weekly = pd.DataFrame(rows)


    
except Exception as e:
    print(e)
```

```
mgl_weekly['Date'] = pd.to_datetime(mgl_weekly['Date'])
```

```
mgl_weekly['Date'] = mgl_weekly['Date'].apply(lambda x: x.strftime('%Y-%m-%d'))
```

```
mgl_weekly.to_csv('/datasets/webscrape/mgl_weekly_'+ datetime.now().strftime('%m_%d_%y.csv'), index_label = 'id')
```
